# Analysis of the Haberman data

```python
import pandas as pd

data = pd.read_csv('./haberman.csv', names=['age', 'operation_year', 'nodes', 'survival'])
data['survival'] -= 1

print(data.describe())
```

### Is there a difference in survival rates when comparing young patients to older patients?

I'll consider young patient as a person who is younger than 52 years old - younger than the mean (and median).


```python
import matplotlib.pyplot as plt
import scipy

plt.scatter(data['age'], data['survival'], marker='x')
plt.xlabel('age')
plt.ylabel('survival')
plt.show()

print('Survival given young or old pateint: ttest:%s, pvalue:%s' % scipy.stats.ttest_ind(data[data['age'] <= 52]['survival'], data[data['age'] > 52]['survival'], axis=0))
```

As we can see there is not much difference in survival rates when comparing young patients to older patients.

### Have survival rates improved over the years?

```python
data_by_operation_year = data.groupby('operation_year').mean()
data_by_operation_year.reset_index(level=0, inplace=True)
data_by_operation_year.plot.bar('operation_year', 'survival', legend=False, title='percentage of people died within 5 years by the year of operation')
plt.show()
```

```python
from scipy import stats

operation_year = np.array(data_by_operation_year['operation_year'])
ranked_years = stats.rankdata(operation_year)

print('pearson correlation between year and survival: r=%s p=%s' % stats.pearsonr(operation_year, data_by_operation_year['survival']))
print('spearman correlation between year and survival: r=%s p=%s' % stats.spearmanr(operation_year, data_by_operation_year['survival']))
print('kendall correlation between year and survival: r=%s p=%s' % stats.kendalltau(ranked_years, data_by_operation_year['survival'].rank()))
```

There is no correlation between operation year and survival rates.

### Which was the best year? Is it statistically different from #2? From the worst one?

I assume that the "best year" means the lowest percentage of people who died within 5 years by the year of operation.

As we can see from the plot above:
1. 1961 was the best year
2. 1960 was the second best year
3. 1965 was the worst one

```python
print('Survival rate of 1961 vs 1960: ttest=%s p=%s' % scipy.stats.ttest_ind(data[data['operation_year'] == 61]['survival'], data[data['operation_year'] == 60]['survival']))
print('Survival rate of 1961 vs 1965: ttest=%s p=%s' % scipy.stats.ttest_ind(data[data['operation_year'] == 61]['survival'], data[data['operation_year'] == 65]['survival']))
```

1961 is not statistically different from 1962, however it is different from 1965.

### Does the number of LNMs affect survival? Maybe it did in certain years and did not in others?

```python
def print_corr(year, nodes, survival):
    print('pearson correlation between nodes and survival for {}:\nr={}\tp={}'
          .format(year, *scipy.stats.pearsonr(nodes, survival)))
    print('spearman correlation between nodes and survival for {}:\nr={}\tp={}'
          .format(year, *scipy.stats.spearmanr(nodes, survival)))
    print('kendall correlation between nodes and survival for {}:\nr={}\tp={}'
          .format(year, *scipy.stats.kendalltau(nodes, survival)))
    
data_by_nodes = data.groupby('nodes').mean()
data_by_nodes.reset_index(level=0, inplace=True)
data_by_nodes.plot.bar('nodes', 'survival', legend=False, title='percentage of people died within 5 years by the number of LNMs nodes')
plt.show()
print_corr('all years', data.nodes, data.survival)
```

```python
operation_years = np.unique(data['operation_year'].values)

for year in operation_years:
    data_for_year = data[data['operation_year'] == year]
    data_by_nodes = data_for_year.groupby('nodes').mean()
    data_by_nodes.reset_index(level=0, inplace=True)
    title = 'percentage of people died within 5 years by the number of LNMs nodes when the operation year was %s' % (1900 + year)
    data_by_nodes.plot.bar('nodes', 'survival', legend=False, title=title)
    plt.show()
    print_corr(1900 + year, data_for_year.nodes, data_for_year.survival)
```

The number of LNMs affects survival.

We can see that there is a correlation between the number of nodes an the survival rates across all years.
<br>By year this is not true. There are some years that the survival rate is distributed almost uniformly with the number of nodes, like 1968 and 1969. However we can tell that in 1960 there is a strong correlation between the number of nodes an the survival rate, with pearson correlation coefficient of 0.614 and a p value of 0.0005.

### Draw a plot of the pdfs for Y1 and Y2 on the same axes, over the x range of [0 10].

```python
from scipy.stats import lognorm, norm
from numpy.core import exp
import numpy as np
import matplotlib.pyplot as plt

distX = lognorm(s=0.25, scale=exp(0))
distY = lognorm(s=1, scale=exp(0))

x = np.linspace(0, 10, 200)
pdfX = distX.pdf(x)
pdfY = distY.pdf(x)
plt.plot(x, pdfX, label='Y1')
plt.plot(x, pdfY, label='Y2')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.legend()

plt.title(r'pdf of $Y1\sim\mathrm{LogNormal}(0,0.25)$ and $Y2\sim\mathrm{LogNormal}(0,1)$')
plt.show()

```

# CLT for Markov chains

### Construct 1000 trajectories, each of length 20.

```python
import random
import numpy as np

transition_matrix = [
    [0.5,  0.25, 0,    0,    0,    0.25],
    [0.25, 0.5,  0.25, 0,    0,    0],
    [0,    0.25, 0.5,  0.25, 0,    0],
    [0,    0,    0.25, 0.5,  0.25, 0],
    [0,    0,    0,    0.25, 0.5,  0.25],
    [0.25, 0,    0,    0,    0.25, 0.5]
]

def create_trajectories(n):
    trajectories = []
    for i in range(0, 1000):
        first_roll = random.randint(1, 6)
        rolls = [first_roll]

        for i in range(0, n-1):
            xi = np.random.choice([1, 2, 3, 4, 5, 6], p=transition_matrix[rolls[i] - 1])
            rolls.append(xi)

        trajectories.append(rolls)
    return trajectories

trajectories_20 = create_trajectories(20)
```

#### What do you expect the average value of all 20 numbers in a trajectory to be?
The main diaigonal has the largest probabilty which means that it if the i roll was n the i+1 roll will probably also be n. Therefore the first roll is very important.
<br>However because the sum of probabilities of each dice side is equal (=1) I expect the average to be 3.5.


#### Compute the average value of each such trajectory. Draw a histogram of the 1000 numbers you received, using 100 bins.

```python
from matplotlib import pyplot as plt
import math

def calc_avg(trajectories):
    avg = []
    for i in range(0, 1000):
        avg.append(np.average(np.array(trajectories[i])))

    return avg

def avg_plot(avg):
    bins = np.linspace(math.floor(min(avg)), 
                       math.ceil(max(avg)),
                       100)

    plt.xlim([1, 6])

    plt.hist(avg, bins=bins)
    plt.xlabel('average')
    plt.ylabel('count')

    plt.show()

avg_20 = calc_avg(trajectories_20)
avg_plot(avg_20)
```

#### What does the distribution look like? What are the empirical mean and the std?
The distribution looks like a normal distribution.

```python
print ('Empirical mean = %s' % np.average(avg_20))
print ('Empirical std = %s' % np.std(trajectories_20))
```

### Draw normal fit curves on your two histograms

```python
from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

def fit_curve(data, length):
    # best fit of data
    (mu, sigma) = norm.fit(data)

    # the histogram of the data
    n, bins, patches = plt.hist(data, 60, normed=1)

    # add a 'best fit' line
    y = mlab.normpdf( bins, mu, sigma)
    l = plt.plot(bins, y, 'r--', linewidth=2)

    #plot
    plt.xlabel('Average')
    plt.ylabel('Probability')
    plt.title(r'$\mathrm{trajectorie\ length=%s,}\ \mu=%.3f,\ \sigma=%.3f$' %(length, mu, sigma))

    plt.show()

fit_curve(avg_20, 20)
fit_curve(avg_2000, 2000)
```
